```python
import pandas as pd
import numpy as np
from ast import literal_eval
import matplotlib.pyplot as plt
from IPython.display import display
```

```python
rest = pd.read_pickle("univ/tor_rest.p")
df = pd.read_pickle("tor_rest_features.p")
```

```python
rest.head()
```

```python
df.head()
```

```python
df_attributes = rest['attributes'].apply(pd.Series)
df_attributes
```

```python
BusParking = df_attributes['BusinessParking']
```

```python
BusParking.isna()
```

```python
#get indices in BusinessParking column that are not NaN
parking_index = BusParking.loc[~BusParking.isna()].index
```

```python
len(parking_index)
```

```python
#convert string into dictionary and append to parking df
parking = pd.DataFrame()
for i in parking_index:
        parking = parking.append(literal_eval(BusParking[i]), ignore_index=True)
```

```python
#add index to parking df
parking.index = parking_index
```

```python
#join
parking_df = pd.DataFrame(BusParking).join(parking, on = 'business_id').drop('BusinessParking', axis=1)
```

```python
# Alcohol column 'none': 0, 'full_bar': 1, 'beer_and_wine': 2
df_attributes['Alcohol'].unique()
df_attributes['Alcohol'] = df_attributes['Alcohol'].replace(to_replace = {'none': 0, 'full_bar': 1, 'beer_and_wine': 2})
df_attributes
```

```python
# RestaurantsAttire 'casual': 1, 'dressy': 2, 'formal': 3
df_attributes['RestaurantsAttire'].unique()
df_attributes['RestaurantsAttire'] = df_attributes['RestaurantsAttire'].replace(to_replace = {'casual': 1, 'dressy': 2, 'formal': 3})
df_attributes
```

```python
#Boolean
df_attributes.replace(to_replace = {'True': 1, 'False': 0}, inplace=True)
```

```python
#Ambience
Ambience = df_attributes['Ambience']
Ambience_index = Ambience.loc[~Ambience.isna()].index
```

```python
Ambience_df = pd.DataFrame()
for i in Ambience_index:
        Ambience_df = Ambience_df.append(literal_eval(Ambience[i]), ignore_index=True)
```

```python
Ambience_df.index = Ambience_index
```

```python
Ambience_new_df = pd.DataFrame(Ambience).join(Ambience_df, on = 'business_id').drop('Ambience', axis=1)
```

```python
Ambience_new_df
```

```python
df_attributes.columns
```

```python
df1 = df_attributes.join(Ambience_new_df, on = 'business_id')
```
