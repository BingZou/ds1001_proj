```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display
```

```python
rest = pd.read_pickle("univ/tor_rest.p")
review = pd.read_pickle("univ/tor_reviews.p")
```

```python
display(rest.head())
display(review.head())
```

```python
review.groupby('business_id', as_index=False)[['stars']].mean()
```

```python
stars = review.groupby('business_id', as_index=False)['stars'].agg(['mean', 'std'])
stars['stars_avg'] = stars['mean']
stars['stars_std'] = stars['std']
stars['business_id'] = stars.index
stars.head()
```

```python
join = rest.join(stars[['stars_avg', 'stars_std']], on='business_id')
diff = np.abs(join['stars'] - join['stars_avg']).sort_values(ascending=False)
(diff > 0.25).sum()
```

```python
join.isna().describe()
```

```python
# attributes, hours have null values
```

```python
df = join[['stars_avg', 'stars_std', 'categories', 'latitude', 'longitude', 'postal_code', 'hours']]
```

```python
categories = pd.DataFrame(columns=['count'])
for category_list in rest.categories:
    for category in category_list.split(','):
        cat_strip = category.strip().lower()
        if cat_strip in categories.index:
            categories.loc[cat_strip] += 1
        else:
            categories.loc[cat_strip] = 1
            
categories.head()
```

```python
categories = categories.astype(int)
categories.describe()
```

```python
int(categories.loc[category])
```

```python
df['other_category'] = False

for category in categories.index:
    if int(categories.loc[category]) < 10:
        df['other_category'] = df['other_category'] | df['categories'].str.contains(category)
```

```python
hrs_df = df[~df.hours.isna()]
daily_hrs = hrs_df['hours'].apply(pd.Series)
```

```python
start_end_df = pd.DataFrame(index=hrs_df.index)

for day in daily_hrs.columns:
    
    start_hr = daily_hrs[day].str.split(':').str[0].fillna('12').astype(int)
    start_min = \
        daily_hrs[day].str.split(':').str[1].str.split('-').str[0].fillna('0').astype(int)
    start = start_hr + start_min/60
    start_end_df["%s_start" % day] = start
    
    end_hr = \
        daily_hrs[day].str.split('-').str[1].str.split(':').str[0].fillna('12').astype(int)
    end_min = \
        daily_hrs[day].str.split('-').str[1].str.split(':').str[1].fillna('0').astype(int)
    end = end_hr + end_min/60
    end[end < start] += 24
    start_end_df["%s_end" % day] = end
```

```python
df = df.merge(start_end_df, how='left', left_on=df.index, right_on=start_end_df.index)
```

```python
df.fillna(df.mean(), inplace = True)
```

```python
df['business_id'] = df['key_0']
df.drop(['key_0', 'hours'], axis=1, inplace=True)
```

```python
df.to_pickle("tor_rest_features.p")
```
