```python
df_dine_small = dfdine[['ROW_ID', 'ESTABLISHMENT_STATUS', 'SEVERITY']].copy()
df_dine_small.head()
```

```python
dfm = dfrest.join(dfmap, how='left').reset_index().merge(df_dine_small, how='left', left_on='mapped_ROW_ID', right_on='ROW_ID')
```

```python
df_final = dfm[['business_id', 'ESTABLISHMENT_STATUS', 'SEVERITY']].copy()
df_final.head()
```

```python
df_final
```

```python
es = pd.get_dummies(dfm['ESTABLISHMENT_STATUS'])
sev = pd.get_dummies(dfm['SEVERITY'])
```

```python
df_final.columns
```

```python
renamer = {
    'Closed':'dinesafe_rev_closed',
    'Conditional Pass':'dinesafe_rev_condpass',
    'Pass':'dinesafe_rev_pass',
    'C - Crucial':'dinesafe_status_crucial',
    'M - Minor':'dinesafe_status_minor',
    'NA - Not Applicable':'dinesafe_status_na',
    'S - Significant':'dinesafe_status_significant',
}
```

```python
df_final = pd.concat([dfm[['business_id']], es, sev], axis=1)
df_final.rename(columns=renamer, inplace=True)
df_final.set_index('business_id', inplace=True)
df_final.head()
```

```python
df_final.to_pickle('features/dinesafe.p')
```
