```python
import pandas as pd
import numpy as np
from ast import literal_eval
import matplotlib.pyplot as plt
from IPython.display import display
from datetime import datetime
```

```python
users = pd.read_pickle("univ/tor_users.p")
```

```python
users
```

```python
#elite
elite = users['elite']
elite_notna_index = elite.loc[~elite.isna()].index
```

```python
#replace elite column with number of years having elite status
elite_df = []
for i in elite_notna_index:
    elite_df.append(elite[i].count(",")+1)
elite_df = pd.DataFrame(elite_df, index = elite_notna_index, columns = ['number of years of elite'])
```

```python
users = users.join(elite_df, on = 'user_id')
```

```python
users.drop('elite', axis=1, inplace = True)
users
```

```python
#friends
friends = users['friends']
friends_notna_index = friends.loc[~elite.isna()].index
```

```python
#replace friends column with number of friends
friends_df = []
for i in friends_notna_index:
    friends_df.append(friends[i].count(",")+1)
friends_df = pd.DataFrame(friends_df, index = friends_notna_index, columns = ['number of friends'])
```

```python
users = users.join(friends_df, on = 'user_id')
```

```python
users.drop('friends', axis=1, inplace = True)
```

```python
#drop name column
users.drop('name', axis=1, inplace = True)
users
```

```python
#years since replace yelping_since with membership time
#ref https://engineeringblog.yelp.com/2018/08/yelp-dataset-challenge-round10-winner.html
membertime = []
release_date = datetime.strptime('2018-08-02', "%Y-%M-%d")
yelping_since = users['yelping_since']
for i in range(len(users['yelping_since'])):
    membertime.append((release_date - datetime.strptime(yelping_since[i], "%Y-%M-%d")).days)
membertime_df = pd.DataFrame(membertime, index = users.index, columns = ['membershiptime'])
```

```python
users = users.join(membertime_df, on = 'user_id')
```

```python
users.drop('yelping_since', axis=1, inplace = True)
```
